```python
path = './data/person_t/chr1/'
g = os.walk(path)  


for path,dir_list,file_list in g:  
    for file_name in file_list:
        with open(path+file_name,'r+') as f:
            person = file_name.split('.')[0][:-2]
            lines = f.read()
            f.seek(0,0)
            f.write(person+lines)
        
        
```

```python
path = './data/person_t/'
g = os.walk(path)  
file_name1 = 'SC_GMWOF5428906\n_A.txt'
del_name1 = 'SC_GMWOF5428906_A.txt'
file_name2 = 'SC_GMWOF5428906\n_B.txt'
del_name2 = 'SC_GMWOF5428906_B.txt'
for path,dir_list,file_list in g:  
    for dir_name in dir_list:
        if(dir_name != 'chr1'):
            print(dir_name)
            person = 'SC_GMWOF5428906'
            os.remove(path+dir_name+'/'+del_name1)
            with open(path+dir_name+'/'+file_name1,'r+') as f:
                lines = f.read()
                f.seek(0,0)
                f.write(person+lines)
            os.rename(path+dir_name+'/'+file_name1, path+dir_name+'/'+del_name1)
            os.remove(path+dir_name+'/'+del_name2)
            with open(path+dir_name+'/'+file_name2,'r+') as f:
                lines = f.read()
                f.seek(0,0)
                f.write(person+lines)
            os.rename(path+dir_name+'/'+file_name2, path+dir_name+'/'+del_name2)
```

```python
def get(name, info):
    # 写入新文件夹
    with open('./data/person_all/'+name, "a") as f:
        f.write(''.join(info))
        f.write('\n')
```

```python
result = pd.DataFrame(columns = ('name','len'))
for path,dir_list,file_list in os.walk('./data/person_all/'): 
    for file_name in file_list:
        result = result.append(pd.DataFrame({'name':[file_name],'len':[len(open('./data/person_all/'+file_name,'rU').readlines())]}))
```

```python
result = result.sort_values(by = ['name'])
result.to_csv('./data/result.csv',index = False)
```
